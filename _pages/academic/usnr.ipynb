```r
library(dplyr)
library(stringr)
df = read.csv('usnr.csv')
df %>% head
```

```r
library(stargazer)
df2 <- df %>% filter(str_detect(University, 'Bio.|Public|Health')) %>% 
arrange(Rank) %>% 
mutate(Rank2 = rank(Rank, ties.method = 'min')) %>%
select(-X)
stargazer(df2, out = 'usnr_stat.html', type = 'html', summary = F)
```
